# Common pool problem

The common pool problem is a game where the players have to decide how much to take from a common pool. Each player has the following utility function:
$$
U_i[t] = E_t\left(\sum_{s=0}^\infty \beta^s \log (c^i_{t+s})\right)
$$
, where $c^i_{t+s}$ is the consumption of player $i$ at time $t+s$. The society has the following common ressource $K_t$:
$$
K_{t+1} = A \cdot K_t - \sum_i^N c^i_t
$$

$A$ must be bigger than 0, but it is allowed to be larger than 1. The rules are the following:
- All players announce their consumption at time $t$ at the same time.
- If the sum of all announced consumptions is smaller than the common pool, then the announced consumptions are allowed.
- If the sum of all announced consumptions is larger than the common pool, then the whole pool is distributed.

## Single person decision

Suppose we have a single person who has to decide how much to take from the common pool. The player will take the same amount for all the players in the society. The player has to decide how much to take from the common pool. The player has the following utility function:
$$
U[0] = E_0 \sum_{s=0}^\infty \beta^s \log (C_{t})
$$,
where $C_t$ is the total consumption of the society at time $t$. 


## The fishery problem

We name the annual catch $y$. It depends on the stock of fish $s$ and the number of fishing boats $b$. The dependency between the different variables is given by the following equation:

- $y = f(s,b)$. The catch depends on the stock and the number of boats.
- $\frac{\partial y}{\partial s} > 0$, the catch increases with the stock.
- $\frac{\partial y}{\partial b} > 0$, the catch increases with the number of boats.
- $\frac{\partial^2 y}{\partial b^2} < 0$, each additional boat catch less.

We can finally write the cost of fishing as $c = wb$, where $w$ is the cost of sending out a single boat. Let us now sketch the cost and the revenue of fishing as a function of the number of boats.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f, ax = plt.subplots(1, 1, figsize=(10, 5))

b = np.linspace(0,2,100)
w = 3
c = w*b
y = np.sqrt(10*b)

ax.plot(b, c, label='Cost')
ax.plot(b, y, label='Revenue')
ax.set_xlabel('Number of boats')
ax.set_title('Cost and revenue of fishing')
```
